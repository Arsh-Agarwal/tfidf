# Link Grabbers

```python
def getLC_ATags(url):
    driver.get(url)
    time.sleep(7)

    ## supposedly page load completed
    links = driver.find_elements(By.TAG_NAME, "a")

    valid = []
    pattern = '/problems/'
    fbpattern = '/solution'
    for link in links:
        href = link.get_attribute("href")
        if href == None: continue
        if href[-9:]==fbpattern: continue
        if pattern in href:
            valid.append(href)
            print(href)
    return valid

def getCF_ATags(url):
    print("Page load init... url: ", url)
    driver.get(url)
    print("Waiting for page to load... ")
    time.sleep(5)
    print("grabbing links...")

    ## supposedly page load completed
    links = driver.find_elements(By.TAG_NAME, "a")

    valid = []
    pattern = '/problem/'
    fbpattern = '/status/'
    for link in links:
        href = link.get_attribute("href")
        if href == None: continue
        if fbpattern in href: continue
        if pattern in href:
            valid.append(href)
            print(href)
    return valid
```

## Drivers

### Leetcode

```python
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service=service)

opfile = open('leetcode.txt', 'w')
links = []
for i in range(1, 58):
    url = lcBaseurl+str(i)
    links += getLC_ATags(url)
    print("done: " + url)

links = set(links)
for link in links: opfile.write(link+'\n')

opfile.close()
```

### Codeforces

```python
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service=service)

opfile = open('codeforces.txt', 'w')
links = []
for i in range(1, 2):
    url = cfBaseurl+str(i)
    print("calling grabber function...")
    getCF_ATags(url)
    links += getCF_ATags(url)
    print("done: " + url)

links = set(links)
for link in links: opfile.write(link+'\n')

opfile.close()
```

## Leetcode

```python
service = Service('/home/murphy/tfidf/chromedriver')
driver = webdriver.Chrome(service=service)
heading_class = ".mr-2.text-label-1"
body_class = "._1l1MA"
index_file = 'index.txt'
urlindex_file = 'urlindex.txt'
datafolder = 'qdata/'

def getLCPageData(url, index):
    try:
        driver.get(url)
        time.sleep(7)
        heading = driver.find_element(By.CSS_SELECTOR, heading_class)
        body = driver.find_element(By.CSS_SELECTOR, body_class)
        print(heading.text)
        print(body.text)
        if(heading.text):
            with open(index_file, 'a') as file: file.write(heading.text + '\n')
            with open(urlindex_file, 'a') as file: file.write(url+'\n')
            
            folderpath = datafolder + str(index)
            os.makedirs(folderpath, exist_ok=True)
            filepath = folderpath + '/' + str(index) + '.txt'
            with open(filepath, 'w') as file: file.write(body.text)
        time.sleep(1)
        return True
    except Exception as e:
        print("error: ", e)
        return False
```

## Codeforces

```python
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service=service)
heading_class = ".problem-statement"
index_file = 'index.txt'
urlindex_file = 'urlindex.txt'
datafolder = 'qdata/'

def getCFPageData(url, index):
    try:
        driver.get(url)
        time.sleep(7)
        heading = driver.find_element(By.CSS_SELECTOR, heading_class)
        bodyparent = driver.find_element(By.CSS_SELECTOR, '.problem-statement')
        body = bodyparent.find_elements(By.CSS_SELECTOR, 'div')
        tot = len(body)
        bodytext = ""
        for i in range(10, tot):
            text = body[i].text
            if(len(text) > 5 and text[:6]=='Input\n'): break
            bodytext += text
        
        if(heading.text):
            with open(index_file, 'a', encoding='utf-8') as file: file.write(heading.text + '\n')
            with open(urlindex_file, 'a', encoding='utf-8') as file: file.write(url+'\n')
            
            folderpath = datafolder + str(index)
            os.makedirs(folderpath, exist_ok=True)
            filepath = folderpath + '/' + str(index) + '.txt'
            with open(filepath, 'w', encoding='utf-8') as file: file.write(bodytext)
        time.sleep(1)
        return True
    except Exception as e:
        print("error: ", e)
        return False
```

```python
links = []
with open('leetcode.txt', 'r') as file:
    for line in file:
        if line!='\n': links.append(line)

index = 1
for link in links:
    if getLCPageData(link, index): index+=1

# links = []
# with open('codeforces.txt', 'r') as file:
#     for line in file:
#         if line!='\n': links.append(line)

# for link in links:
#     if getCFPageData(link, index): index+=1
```
